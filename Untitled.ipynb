```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
# import matplotlib.ticker as mticks

import missingno as miss

import seaborn as sns
import numpy as np
```

```python
sns.set_context('notebook', font_scale=0.9)
plt.style.use(['ggplot'])
```

```python
weather = pd.read_csv("london_weather_MODIFIED - london_weather_MODIFIED.csv")
weather.head()
```

```python
weather.shape
```

**Convert Datetime column to datetime type with pd.to_datetime**

```python
## Compare to dates in dataframe
weather['date'][0]
```

```python
weather['date'] = pd.to_datetime(weather['date'], format='%Y%m%d')
```

```python
weather.dtypes
```

```python
##Set Date as index
weather = weather.set_index('date')
```

```python
weather.index
```

```python
##Keep only data from the year 2000 or later
weather = weather.loc['2000': ]
weather
```

Keep only the following features:
precipitation, 
mean_temp, 
min_temp, 
max_temp, 
snow_depth

```python
weather = weather.drop(['global_radiation', 'pressure', 'cloud_cover', 'sunshine'], axis=1)
weather.head()
```

**Impute any missing values**

```python
weather.isnull().sum()
```

```python
weather_interp = weather.interpolate()
```

```python
weather_interp.plot()
```

```python
df1 = weather_interp.copy()
df1.head()
```

```python
df2 = weather_interp.copy()
df2.head()
```

**What month had the most precipitation between 2000 through 2010?**

```python
##Slice out 2000 - 2010 
df1 = df1.loc['2000':'2010']
df1.head()
```

```python
df1.index
```

```python
#resample as Months frequency
df1 = df1.resample('M').sum()
df1
```

```python
df1.index
```

```python
df1['precipitation'].plot(figsize=(12,3));
```

```python
## setting the default figsize for this notebook
plt.rcParams['figure.figsize'] = (12,3)
```

```python
month_max_precip = df1['precipitation'].idxmax()
max_precip = df1['precipitation'].max()
print(f"The highest monthly preciptation was {max_precip} inches, which occured in {month_max_precip.strftime('%B, %Y')}")
```

```python
fig, ax = plt.subplots(figsize=(12,3))
```

```python
fig, ax = plt.subplots(figsize=(12,3))
ax = df1['precipitation'].plot(ax=ax)
ax.set( title='Precipitation from 2000-2010',
      ylabel='Inches')
plt.axvline(x='102000', color='black').set_label('Max Precipitation: 152.7 in October 2000')
plt.legend(loc="upper right")
# customize major ticks 
major_loc =  mdates.YearLocator(1)
major_fmt = mdates.DateFormatter("%Y")

ax.xaxis.set_major_locator(major_loc)
ax.xaxis.set_major_formatter(major_fmt)


# customize minor ticks
minor_loc =  mdates.MonthLocator(interval=3)

ax.xaxis.set_minor_locator(minor_loc)

## Final xdate tweaks
fig = ax.get_figure()
fig.autofmt_xdate(which='major',rotation=90)
```

 **Which year between 2000-2020 had the coolest average temperature?**

```python
##Slice out 2000 - 2020
df2 = df2.loc['2000':'2020']
df2.head()
```

```python
#resample as year frequency
df2 = df2.resample('Y').mean()
df2
```

```python
year_min_temp = df2['mean_temp'].idxmin()
min_temp = df2['mean_temp'].min()
print(f"The coolest average temp was {min_temp * .01} degrees, which occured in {year_min_temp.strftime('%Y')}")
```

```python
fig, ax = plt.subplots(figsize=(12,3))
```

```python
ax = df2['mean_temp'].plot()
ax.set( title='Average Temperature',
      ylabel='Degrees')
plt.axvline(x='2010', color='black').set_label('Minimum avg temp: 38.9 degrees in 2010')
plt.legend(loc="upper right")
# customize major ticks 
major_loc = mdates.YearLocator(5)
major_fmt = mdates.DateFormatter("%Y")

ax.xaxis.set_major_locator(major_loc)
ax.xaxis.set_major_formatter(major_fmt)
# customize minor ticks
minor_loc =  mdates.YearLocator(1)

ax.xaxis.set_minor_locator(minor_loc)
## Final xdate tweaks
fig = ax.get_figure()
fig.autofmt_xdate(which='major',rotation=90);
```
